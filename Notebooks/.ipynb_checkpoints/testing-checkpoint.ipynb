```python
con = sq.connect('{}table.db'.format(data_dir))
c = con.cursor()

def create_table(table):
    c.execute("""CREATE TABLE IF NOT EXISTS ?
              (
                  Index REAL
                  Name TEXT
              )
              """, (table))
    
def add_column(table, key, dtype):
    try:
        c.execute("""
            ALTER TABLE table
            ADD COLUMN key dtype
            """)
    except:
        print("Column Exists...")
        pass
    
def data_entry(table, key, value, idx):
    c.execute("""
        INSERT INTO ? 
        (key) 
        VALUES (?) 
        WHERE Name = ?
        """
        (table, key, value, idx))
    
    con.commit()
    #c.close()
    #con.close()
```
