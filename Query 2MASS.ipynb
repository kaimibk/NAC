## Query fp_psc : 2MASS All-Sky Point Source Catalog (PSC)

```python
from astroquery.irsa import Irsa
```

```python
features = [
            "j_m", "j_msigcom", 
            "h_m", "h_msigcom",
            "k_m", "k_msigcom"
           ]
from astropy.coordinates import match_coordinates_sky

def retrieve(idx, df):
    global features
    global c
    temp = Irsa.query_region(c[idx], 
                     radius=5*u.arcsecond,
                    catalog="fp_psc").to_pandas()
    
    catalog = SkyCoord(ra=temp.ra, dec=temp.dec,
                       unit=(u.deg, u.deg))
    
    i,_,_ = match_coordinates_sky(c[idx], catalog)
    
    for feat in features:
        df.at[idx, feat] = temp.at[int(i), feat]
```

```python
[retrieve(i, df) for i, row in df.iterrows()]
```
